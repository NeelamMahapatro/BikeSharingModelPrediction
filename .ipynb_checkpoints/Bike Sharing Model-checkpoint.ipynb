# Bike Sharing Model building, Prediction and Evaluation

## 1. Importing Libraries and Reading Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import statsmodels
import statsmodels.api as sm
import sklearn
from sklearn.model_selection import train_test_split

from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
bikeSharingData = pd.read_csv("day.csv")
bikeSharingData.head()
```

```python
print(bikeSharingData.info())
print(bikeSharingData.shape)
```

* No empty cells are detected

```python
bikeSharingData.describe()
```

## 2. Data Cleaning and EDA 

```python
# Dropping dteday column since Yr column is there for representing the year
bikeSharingData = bikeSharingData.drop(["dteday"],axis=1)

# Dropping casual and registered column since cnt column is there which is sum of both these column values
bikeSharingData = bikeSharingData.drop(["casual", "registered"],axis=1)

# Dropping holiday column since working day column is already there and it seems duplicate column
bikeSharingData = bikeSharingData.drop(["holiday"],axis=1)

# Dropping instant column since its for indexing only purpose
bikeSharingData = bikeSharingData.drop(["instant"],axis=1)

bikeSharingData.head()
```

### Encoding of Categorical Variable 

* weathersit - 4 levels. So we need to create 4-1=3 dummy variable.
    1. weathersit_1: clear
    2. weathersit_2: Mist + cloudy
    3. weathersit_3: Light Snow
* season - 4 levels. So we need to create 4-1=3 dummy variable.
    1. season_1: spring
    2. season_2: summer
    3. season_3: fall
    4. season_4: winter
* month - 12 levels
    - 1 - January, 2 - February and so on.

```python
sns.boxplot(x = 'season', y = 'cnt', data = bikeSharingData)
plt.show()
sns.boxplot(x = 'weathersit', y = 'cnt', data = bikeSharingData)
plt.show()
sns.boxplot(x = 'mnth', y = 'cnt', data = bikeSharingData)
plt.show()
```

```python
# Encoding Categorical Column Season   
bikeSharingData = pd.get_dummies(bikeSharingData,columns=['season', 'weathersit', 'mnth'], drop_first=True)
bikeSharingData.head()
```

## 3. Data Visualization - (Seaborn and Matplotlib)

#### PairPlot to analyze different variables

```python
bikeSharingNumericCols = ["cnt","temp","atemp","hum","windspeed"]
bikeSharingNumericData = bikeSharingData[bikeSharingNumericCols]
sns.pairplot(bikeSharingNumericData)
plt.show()
```

```python
# yr = 1 is for 2019 and 0 for 2018
sns.boxplot(x = 'yr', y = 'cnt', data = bikeSharingData)
```

```python
sns.boxplot(x = 'weekday', y = 'cnt', data = bikeSharingData)
```

#### From above boxplot, we can observe
    1. features like season, year, month, weathersit has significant effect on count where as
    2. features like holiday, weekday, workingday are not having much significant effect on count
    3. temp and atemp are highly correlated, so we can remove atemp feature

```python
bikeSharingData = bikeSharingData.drop(["atemp"],axis=1)
```

```python
bikeSharingData.describe()
```

## 4. Model Building - Linear Regression

-  Splitting data into train test split

```python
np.random.seed(0)
bikeSharing_train, bikeSharing_test = train_test_split(bikeSharingData, train_size = 0.7, test_size = 0.3, random_state = 100)
```

* Rescaling the Features
   - Rescaling the variables to have a comparable scale.
   - Applying min-max scaling to winspeed, hum and temp

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
features = ['temp', 'hum', 'windspeed']
prediction = ['cnt']

bikeSharing_train[features] = scaler.fit_transform(bikeSharing_train[features])
bikeSharing_train[prediction] = scaler.fit_transform(bikeSharing_train[prediction])
```

```python
bikeSharing_train.head()
```

- Using of LinearRegression function of SciKit Learn. It is compatibile with RFE (Recursive feature elimination)
* Here we will use RFE to detect the 10 most important features, then we will use manual backward elimination for discarding the features one by one

```python
# importing RFE module
from sklearn.feature_selection import RFE

y_train = bikeSharing_train.pop('cnt')
X_train = bikeSharing_train
linearRegModel = LinearRegression()
linearRegModel.fit(X_train, y_train)
```

```python
rfe = RFE(linearRegModel)
rfe = rfe.fit(X_train, y_train)
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
# selecting 12 features 
rfeSupportcols = X_train.columns[rfe.support_]
rfeSupportcols
```

- Building model using statsmodel

```python
X_train_VIF = X_train[rfeSupportcols] 
X_train_LM = sm.add_constant(X_train_VIF)
linearModel1 = sm.OLS(y_train,X_train_LM).fit() 
```

```python
# Summary of the Model
print(linearModel1.summary())
```

```python
def calculateVIF(xData):
    VIF1 = pd.DataFrame()
    VIF1['Features'] = xData.columns
    VIF1['VIF'] = [variance_inflation_factor(xData.values, i) for i in range(xData.shape[1])]
    VIF1 = VIF1.sort_values(by = "VIF", ascending = False)
    return VIF1
calculateVIF(X_train_VIF)
```

- Eliminating humidity column based on its high VIF value

```python
# Model 2 eliminating Humidity
X_train_VIF = X_train_VIF.drop(["hum"], axis = 1)
X_train_LM = sm.add_constant(X_train_VIF)
linearModel2 = sm.OLS(y_train,X_train_LM).fit() 
print(linearModel2.summary())
print(calculateVIF(X_train_VIF))
```

- Dropping mnth_10 for its higher p-value
- temp and windspeed seems to be an important factor even if with high VIF and hence we are not dropping them

```python
# Model 3 eliminating season_4
X_train_VIF = X_train_VIF.drop(["mnth_10"], axis = 1)
X_train_LM = sm.add_constant(X_train_VIF)
linearModel3 = sm.OLS(y_train,X_train_LM).fit() 
print(linearModel3.summary())
print(calculateVIF(X_train_VIF))
```

### Final Model Interpretation
    * Hypothesis Testing:

        H0: All the slope intercepts are zero i.e B1=B2=...=Bn=0
        H1: at least one Bi!=0
        
    * From the Model we have coefficient values as below-
    

```python
# Printing Coefficients of the Final Model
finalLinearModel = linearModel3
finalLinearModel.params
```

#### 1. Positive coefficients (m) indicates that a unit increase in that feature increases the overall bike hire counts by m units.
    Example - temperature, year
#### 2. Negative coefficients (-m) indicates that a unit increase in that feature decreases the overall bike rental count by m units.
    Example - windspeed, weathersit_2

## 5. Residual Analysis 

```python
y_train_pred = finalLinearModel.predict(X_train_LM)
residual = y_train-y_train_pred
sns.distplot((residual), bins = 20)
plt.show()
```

#### - Residuals are normally distributed and hence its valid linear model

## 6. Prediction Using Final Linear Model 

#### Preparing Test data

```python
bikeSharing_test[features] = scaler.fit_transform(bikeSharing_test[features])
bikeSharing_test[prediction] = scaler.fit_transform(bikeSharing_test[prediction])
bikeSharing_test.head()
```

```python
y_test = bikeSharing_test.pop('cnt')
X_test = bikeSharing_test
x_TrainCols = X_train_VIF.columns
X_test=X_test[x_TrainCols]
# Adding constant variable to test dataframe
X_test_LM = sm.add_constant(X_test)
X_test_LM.info()
```

```python
# Prediction on Test data
y_test_pred = finalLinearModel.predict(X_test_LM)
plt.scatter(y_test, y_test_pred, alpha=.5)
plt.show()
```

## 7. R2 and Adjusted R2 Value for Test

```python
r2 = r2_score(y_test, y_test_pred)
print("r2 score for Test Model is: ", r2)

adj_r2 = 1-(1-r2)*(X_test.shape[0]-1)/(X_test.shape[0]-X_test.shape[1]-1)
print("Adjusted r2 score for Test Model is: ", adj_r2)
```
